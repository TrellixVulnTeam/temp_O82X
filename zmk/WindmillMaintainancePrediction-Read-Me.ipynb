<h1>Windmill Maintenance Prediction Usecase</h1>

Windmill Prediction Maintenance uses LSTM (Long Short Term Memory) Model with features such as Rotation, MeanPressure, MeanVibration, ErrorCount3, ErrorCount4, RotationSD, PressureSD, VibrationSD, ErrorCount5, Humidity, GroundWindSpeed, ExternalTemperature, RotationSpeed, ErrorCount6, ErrorCount7, ErrorCount8, VoltageSD_12H, VoltageSD_24H, RUL, Cycle_Norm. It takes 15 records for each features and scores breakdown probability to predict Windmill Maintainance Prediction.

### To jump start, Use UMOYA (Repo).
UMOYA does versioning resources (like model, data and code) and manages its dependencies for ai projects.
MLW's Solution Team publishes Data, Model and Code for this Usecase on UMOYA (Repo) so User(s) can easily grab them all in one click.

<br/></br> There are two ways to download resource(s).
1. Download resource(s) from an MLW tenant UI.
<ol>
    <li>Go to "Repo" Tab.</li>
    <li>Select "WindMillMaintainacePrediction.ipynb" </li>
    <li>Click on "Download" </li>
</ol>
2. Use UMOYA CLI.
Go to MLW/ZMOD resource directory in terminal, Give command : <br/>umoya add WindMillMaintainacePrediction.ipynb

### Test Model :
1. Load "WindMillMaintainacePrediction.pmml" to ZMK.
2. Select "WindMillMaintainacePrediction-Test.csv" in "Data" Tab.
3. Click on "Predect Data".
4. Selct Model "WindMillMaintainacePrediction.pmml" and submit.

It will create new file with column "target" and value of maintainace probability from 15th records.
